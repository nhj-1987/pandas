```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

%matplotlib inline 

mpl.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=["a", "b"])
df["b"] = df["b"] + np.arange(1000) # index 값추가
df.plot.hexbin(x="a", y="b", gridsize=5);
```

```python
df["z"] = np.random.uniform(0, 3, 1000)
df["z"]
```

```python
df.plot.hexbin(x="a", y="b", C="z", reduce_C_function=np.median, gridsize=20); # 평균, 최대값 , 값의 많고 적음 표현 가능 , 산점도 기능 보완
```
