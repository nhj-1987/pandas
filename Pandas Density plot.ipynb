```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#한글 폰트 사용시 그래프에서 마이너스 폰트 깨지는 문제에 대한 대처
plt.rcParams['axes.unicode_minus'] = False
```

```python
ser = pd.Series(np.random.randn(1000))
ser
```

```python
ser.plot.kde(); #kde = density 같은 그래프 , 확률밀도함수 가우시안 정규분포
```

```python
ser.plot.density(); # 커널밀도함수 : 적분값이 1인 함수
```

```python
ser.plot.hist() # 히스토그램
```
