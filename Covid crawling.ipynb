```python
#필요한 라이브러리

import pandas as pd
```

```python
#크롤링으로 가져올 url
url = "C://Users/User/Downloads/seoulcovid.html"
url
```

```python
# table 태그 찾음
table = pd.read_html(url)
len(table)
```

```python
#read_html 사이트내 html 태그를 읽어 확진자 현황을 가져옴
df = table[3]
df
```

```python
#마지막 확진일 기준으로 csv 파일 생성
last_day = df.loc[0,'확진일']
last_day = last_day.replace("-","_")
last_day
```

```python
file_name = f"seoul-covid19_{last_day}.csv"
file_name
```

```python
df.to_csv(file_name,index=False)# 엑셀에서 확인 ,encoding="cp949")
```
